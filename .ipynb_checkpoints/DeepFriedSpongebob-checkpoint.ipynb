```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy import ndimage
%config InlineBackend.figure_format = 'retina'
```

```python
im = io.imread('fried/immaheadout.jpg')
im.shape, im.dtype
```

```python
plt.imshow(im);
```

```python
spangborb =  np.array(im[:,:,0])
plt.imshow(spangborb, cmap='gray');
```

```python
kernel2 = np.ones((10,10)) / 100
spangborb_blur = ndimage.convolve(spangborb, kernel2)

plt.imshow(spangborb_blur, cmap='gray');
```

```python
x = np.arange(-10, 11)

kernel1d = st.norm.pdf(x, 0, .37)

row = kernel1d.reshape((1,-1))  # 1 x 21
col = kernel1d.reshape((-1,1))  # 21 x 1

kernel2d = col @ row  # 21 x 21
```

```python
deep_fried_spongebob = ndimage.convolve(spangborb, kernel2d)

plt.imshow(deep_fried_spongebob, cmap='jet');
```

```python
edge_kernel = np.array([[-1, -1, -1], [-1, +7, -1], [-1, -1, -1]])
#edge_kernel2 = np.array([[-1, -1, -1], [-1, +6, -1], [-1, -1, -1]])


spangborb_edges = ndimage.convolve(deep_fried_spongebob, edge_kernel)
#super_fried = ndimage.convolve(spangborb_edges, edge_kernel2)

plt.imshow(spangborb_edges, cmap='jet');
```
